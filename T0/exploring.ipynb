```python
# Colinha

meses = df.iloc[:,1:-3]

for mes in meses.columns:
    print(sorted(meses[mes]))
```

```python
- ranking das cidades por numero de homicidios por habitante
```

```python
# Somando colunas
agregado_municipio = pd.concat([df.iloc[:,0], meses.sum(axis=1), df.iloc[:,-1]], axis=1)
display(agregado_municipio)
# agregado_municipio[0]
agregado_municipio[0] = agregado_municipio[0] / agregado_municipio["Populacao"] * 1000
display(agregado_municipio)
```

# Distribuição do número de homicidios por habitante por cidade

```python
df_copy = df.copy()

df_copy.iloc[:, 1:-3] = df_copy.iloc[:, 1:-3].div(df_copy["Populacao"], axis = 0).multiply(1000)

for i in range(len(df_copy.index)):
    print(f"Cidade: {df_copy["Municipio"].iloc[i]}")
    print(df_copy.iloc[i,1:-3].quantile([.25, .5, .75]))

df_copy
```
